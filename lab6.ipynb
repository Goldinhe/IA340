```python
import boto3
from botocore.exceptions import ClientError
import json

def get_secret(secret_name):
    region_name = "us-east-1"

    # Create a Secrets Manager client
    session = boto3.session.Session()
    client = session.client(
        service_name='secretsmanager',
        region_name=region_name
    )

    try:
        get_secret_value_response = client.get_secret_value(
            SecretId=secret_name
        )
    except ClientError as e:
        raise e

    secret = get_secret_value_response['SecretString']
    
    return json.loads(secret)
```

```python
import matplotlib.pyplot as plt

# Calculate the year over year growth rate for each state
df.sort_values(['name', 'year'], inplace=True)
df['growth_rate'] = df.groupby('name')['income'].pct_change()

# Pivot the data to get years as index and states as columns
pivot_df = df.pivot(index='year', columns='name', values='growth_rate')

# Plot the data
plt.figure(figsize=(15,8))
for column in pivot_df.columns:
    plt.plot(pivot_df.index, pivot_df[column], marker='', label=column)

plt.legend(loc='best')
plt.title('Year-over-Year Income Growth Rate per State')
plt.xlabel('Year')
plt.ylabel('Growth Rate')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Convert year to datetime format
df['year'] = pd.to_datetime(df['year'], format='%Y')

# Sort the dataframe by name and year
df.sort_values(by=['name', 'year'], inplace=True)

# Calculate the year-over-year income growth rate
df['YoY_growth'] = df.groupby('name')['income'].pct_change()

# Pivot the dataframe to make each state a column
pivot = df.pivot(index='year', columns='name', values='YoY_growth')

# Plot the results
plt.figure(figsize=(10,5))
for column in pivot.columns:
    plt.plot(pivot.index, pivot[column], label=column)

plt.legend(bbox_to_anchor=(1,1), loc="upper left")
plt.title('Year-over-Year Income Growth Rate per State')
plt.xlabel('Year')
plt.ylabel('Growth Rate')
plt.show()
```
